# Step 3 : DETECT MISSING DATA
Find out which values are empty or incomplete in your dataset.

```python
# @title

print ("Null values in the columns of the dataset")
nullcolValue = df.isna().sum()
nullcolValue
null_counts = df.isna().sum().reset_index()
null_counts.columns = ['Column Name', 'Total null values']
null_counts
```

# Step 4 : TREAT MISSING VALUES
Choose whether to remove gaps or replace them with meaningful values.

# WHEN TO USE
If nulls are affecting calculations, models, or reports.

# EXAMPLE
df['age'].fillna(df['age'].median(),inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)

```python
# @title
df['age'].fillna(df['age'].median(),inplace=True)
# When you use inplace=True: Direct Modification: Pandas goes into the memory where df['age'] is stored
# and updates the values right there.
# inplace=False (Default) A new Series/DataFrame with the fix
df['embarked'].fillna(df['embarked'].mode()[0],inplace=True)
```

# Step : 5 DETECT DUPLICATE RECORDS
Find rows that appear more than once in your dataset.

# WHEN TO USE

# EXAMPLE

df.duplicated().sum()


```python
# @title
df.shape
duplicate_count = df.duplicated().sum()
print(f"Total Duplicate Rows: {duplicate_count}")
```

```python
# @title
# Show the duplicates AND their original counterparts
all_duplicates = df[df.duplicated(keep=False)]
print(all_duplicates.sort_values(by=list(df.columns)))
```
