Below program shows how to define global ans local variables in the python program.

```python
#Global Variables
x = "awesome" # global

def myfunc():
  x = "fantastic" # local
  print("Python is " + x)

myfunc()
print("Python is " + x)
```

The Variables which are created outside of a function are known as global variables.The scope of access of them is throughtout program.Global variables can be used inside and outside of function.Global variables behave like local if they are overwritten inside function.
In the below fuction we are assigning value to "**awesome**" initially. Inside the function we are reassigning it to "**fantastic**". When the function is called in the main program (outside myfunct function). It will print the value "awesome".After accessing we are overwritting to **'awesome overwritten again'**.

```python
x='awesome'
print(x)
def myfunc():
  global x
  x = "fantastic"
  print(x)

myfunc()

print("Python is " + x)
x='awesome overwritten again'
print(x)
myfunc()
```
