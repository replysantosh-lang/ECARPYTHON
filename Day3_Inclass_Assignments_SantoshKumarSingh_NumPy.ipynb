Importing Data
np.loadtxt () is a NumPy function used to load data from a text file (like .txt or .csv) into a NumPy array.
It expects numerical data (numbers only) by default

```python
# Import NumPy using its standard alias, np
import numpy as np 

# Create any array from a .txt file

"""
1 2 3
4 5 6
7 8 9
"""
# Only Numeric data
txt = np.loadtxt('PersonNumeric.txt') 
print(txt)

print()

"""
John 25 72.5
Alice 30 68.0
"""
# This will read all values as strings.
txt = np.loadtxt('PersonString.txt', dtype=str) 
print(txt)

print()
"""
Name Age Height
John 25 72.5
Alice 30 68.0
"""
# If the file has a header (column names)
txt = np.loadtxt('PersonStringHeader.txt', dtype=str, skiprows=1)
print(txt)
```

If your file mixes text and numbers (like names and ages), np.genfromtxt() is more flexible:
Name Age Height
John 25 72.5
Alice 30 68.0

```python
txt = np.genfromtxt('Person.txt', dtype=str, delimiter=',')
print(txt)
```

Skipping the header

```python
txt = np.genfromtxt('Person.txt', dtype=str, delimiter=',', skip_header=1)
print(txt)
```

If the file is comma-separated (CSV format)

Name,Age,Amount
John,25,72.5
Alice,30,68.0
Silent,30,68.0
Noise,30,68.0

```python
# Create any array from a .csv file
csv = np.genfromtxt('Person.csv', delimiter=',', dtype=str)
print(csv)
```
